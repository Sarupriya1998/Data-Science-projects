Load the “sample_superstore.csv”dataset provided on LMS into a DataFrame, using the Matplotlib library, plot the following insights from the dataset:
1.Compare the total sales for each category of products
2.Number of deals in each state of America

```python
import pandas as pd
df=pd.read_csv("SampleSuperstore.csv")
import matplotlib.pyplot as plt
b=df[:9994]
a=plt.bar(b.Category,b.Sales)
plt.ylabel('Sales',fontsize=16)
plt.xlabel('Categories',fontsize=16)
plt.title('Total sales for each category of products',fontsize=16)
print("1.Compare the total sales for each category of products:")
plt.show(a)
a_state_dealings=b.groupby('State')['Sales'].count().sort_values(ascending=False).plot.bar(figsize=(15,5))
plt.ylabel('Deals')
plt.xlabel('American States')
plt.title('Total number of deals in each American State',fontsize=16)
print("2.Number of deals in each state of America:")
plt.show()
```

Load the “sample_superstore.csv” dataset provided on LMS into a DataFrame; using the Seaborn library, plot the following insights from the dataset:
1.Print a correlation matrix between Postal code,Quantity,Sales,Discount,and Profit
2.Plot the correlation matrix using a heatmap

```python
import pandas as pd
df=pd.read_csv("SampleSuperstore.csv")
import matplotlib.pyplot as plt
import seaborn as sns
display(df.head())
corr_mat=df.corr()
corr_mat
plt.subplots(figsize=(12,9))
sns.heatmap(corr_mat)
plt.title('Correlation between Postal code,Quantity,Sales,Discount, and profit',fontsize=16)
plt.show()
```
