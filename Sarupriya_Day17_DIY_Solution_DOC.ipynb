Load the “Netflix_titles.csv”data  into  a  DataFrame  and  plot  the  following insights using the Matplotliblibrary:1.How much of the Netflix content are TV shows or Movies?2.In which country is Netflix the most popular?

```python
import pandas as pd
df=pd.read_csv("netflix_titles.csv")
import matplotlib.pyplot as plt
import numpy as np
b=plt.pie(df['type'].value_counts().sort_values(),labels=['Movie','TV Show'],autopct='%0.1f%%',explode=[0.1,0],colors=['blue','red'],startangle=90)
plt.title('Type of Netflix content')
plt.show()
b=df['country'].value_counts().reset_index()
b=b[b['country']/b['country'].sum()>0.01]
plt.figure(figsize=(14,7))
plt.pie(b['country'],labels=b['index'])
plt.show()


```

Load the “starbucks_drinkMenu_expanded.csv” data into a DataFrame and find out which “Beverage_category” has the highest calories; plot the insights using a bar plot with the help of the Seaborn library.

```python
import pandas as pd
df=pd.read_csv("starbucks_drinkMenu_expanded.csv")
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
plt.subplots(figsize=(15,8))
sns.barplot(y='Beverage_category',x='Calories',data=df)
plt.title('Which of the Starbucks drinks have highest calories')
plt.show()
```
